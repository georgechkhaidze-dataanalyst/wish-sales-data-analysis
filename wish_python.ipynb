```python
df = df.drop(columns=[
    "product_url",
    "product_picture",
    "merchant_profile_picture",
    "merchant_info_subtitle",
    "theme",
    "crawl_month"
])
```

```python
df[[
"rating_five_count",
"rating_four_count",
"rating_three_count",
"rating_two_count",
"rating_one_count"
]] = df[[
"rating_five_count",
"rating_four_count",
"rating_three_count",
"rating_two_count",
"rating_one_count"
]].fillna(0)
```

```python
df.isnull().sum()
```

```python
df["product_color"] = df["product_color"].fillna("Unknown")
```

```python
df["product_variation_size_id"] = df["product_variation_size_id"].fillna("Unknown")
```

```python
df["origin_country"] = df["origin_country"].fillna("Unknown")
```

```python
df["merchant_name"] = df["merchant_name"].fillna("Unknown")
```

```python
df["has_urgency"] = df["urgency_text"].notna().astype(int)
```

```python
df = df.drop(columns=["urgency_text"])
```

```python
import numpy as np

df["rating_five_count"] = df["rating_five_count"].replace(0, np.nan)
df["rating_four_count"] = df["rating_four_count"].replace(0, np.nan)
df["rating_three_count"] = df["rating_three_count"].replace(0, np.nan)
df["rating_two_count"] = df["rating_two_count"].replace(0, np.nan)
df["rating_one_count"] = df["rating_one_count"].replace(0, np.nan)
```

```python
df["rating_five_count"] = df["rating_five_count"].fillna(df["rating_five_count"].mode()[0])
df["rating_four_count"] = df["rating_four_count"].fillna(df["rating_four_count"].mode()[0])
df["rating_three_count"] = df["rating_three_count"].fillna(df["rating_three_count"].mode()[0])
df["rating_two_count"] = df["rating_two_count"].fillna(df["rating_two_count"].mode()[0])
df["rating_one_count"] = df["rating_one_count"].fillna(df["rating_one_count"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["has_urgency_banner"])
```
